```python
matrix = np.array([(1,1),(2,0)])
Q = np.zeros(shape=(2,2))
U = np.zeros(shape = (2,2))
c_1 = matrix[:,0]
c_2 = matrix[:,1]
        
f_1 = c_1
f_2 = c_2 - ((np.dot(c_2, f_1)/(np.linalg.norm(f_1))**2)*f_1)

q_1 = f_1/np.linalg.norm(f_1)
q_2 = f_2/np.linalg.norm(f_2)
        
Q[:,0] = q_1
Q[:,1] = q_2
        
U[0,0] = np.linalg.norm(f_1)
U[0,1] = np.dot(c_2, q_1)
U[1,1] = np.linalg.norm(f_2)

A = np.dot(U, Q)
A
```
